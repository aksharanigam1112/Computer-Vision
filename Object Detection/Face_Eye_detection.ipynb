```
import cv2
import matplotlib.pyplot as plt
```

```
# Read in the cascade classifiers for face & eyes
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
```

```
# create a function to detect faces & eyes
def detect_faces_eyes(img):
    # eye_img = img.copy()

    face_rect = face_cascade.detectMultiScale(image=img,
                                              scaleFactor=1.2,
                                              minNeighbors=15)

    for (x, y, w, h) in face_rect:
        cv2.rectangle(img=img,
                      pt1=(x,y),
                      pt2=(x+w,y+h),
                      color=(255,0,0),
                      thickness=2)


    eye_rect = eye_cascade.detectMultiScale(image=img,
                                              scaleFactor=1.2,
                                              minNeighbors=5)
    for (ix, iy, iw, ih) in eye_rect:
        cv2.rectangle(img=img,
                      pt1=(ix,iy),
                      pt2=(ix+iw,iy+ih),
                      color=(0,255,0),
                      thickness=2)

    return img
```

```
# Reading in the image
img = cv2.imread('eye-face.jpeg')
img_fix = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

```
result = detect_faces_eyes(img_fix)
plt.imshow(result)
```
