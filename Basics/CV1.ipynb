```
import cv2
```

```
img = cv2.imread('hp2.jpg')
img
```

```
# With matplotlib we were reading RGB
# With opencv we are reading BGR
plt.imshow(img)
```

```
img_fix = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_fix)
```

```
# Scale the image to GRAY

img_gray = cv2.imread('hp2.jpg',cv2.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```
# Resize the image
img_new = cv2.resize(img_fix,(1000,400))
plt.imshow(img_new)
```

```
# Resize with ratio
width_ratio = 0.3
height_ratio = 0.3

img = cv2.resize(img_fix,(0,0),img_fix,width_ratio,height_ratio)
plt.imshow(img)
```

```
# Flip on Horizontal Axis
img1 = cv2.flip(img_fix,0)
plt.imshow(img1)
```

```
# Flip on Vertical Axis
img2 = cv2.flip(img_fix,1)
plt.imshow(img2)
```

```
# Flip on Horizontal & Vertical Axis
img3 = cv2.flip(img_fix,-1)
plt.imshow(img3)
```

```
# Change the size of canva

lastImg = plt.figure(figsize=(10,7))
ilp = lastImg.add_subplot(111)
ilp.imshow(img_fix)
```
